```python
import pandas as pd 
import numpy as npp
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df.tail()
```

```python
df.describe ()
```

```python
df.isnull()
```

```python
df.rename(columns = {'per capita income (US$)':'per_capita_income'}, inplace = True) 
```

```python
%matplotlib inline
plt.scatter(x=df.year,y=df.per_capita_income,color='blue',marker='*')
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.year.values.reshape(-1, 1), df.per_capita_income)
```

```python
reg.predict([[2020]])
```
